**Question1**


Write a Python function to find the maximum of three numbers using condition statements.

```python
def max_of_three(a, b, c):
  if a >= b and a >= c:
    return a
  elif b >= a and b >= c:
    return b
  else:
    return c


print(max_of_three(5, 15, 10));

```

**Question 2**

Duc wants to invest in a fixed deposit savings account. He decides to deposit $2,000 in an account that offers an annual interest rate of 5%.



*   Calculate the interest earned by Duc after 4 years
*   Determine the total amount in Duc's account (future value) at the end of the 4 years.


Create the function calculate_interest_and_future_value(principal, rate, time) where



 Parameters:

    principal (float): The initial amount of money.

    rate (float): The interest rate (as a percentage).

    time (float): The time period for which the interest is calculated.

    interest = (principal * rate * time) / 100


```python
#def calculate_interest_and_future_value(principal, rate, time):
def calculate_interest_and_future_value(principal, rate, time):
    interest = (principal * rate * time) / 100
    future_value = principal + interest
    return interest, future_value

# Example usage
interest, future_value = calculate_interest_and_future_value(2000, 5, 4)
print(f"Interest: ${interest}, Future Value: ${future_value}")  # Output: Interest: $400.0, Future Value: $2400.0
```

**Question 3**

Duc is deciding between two job offers. One job pays an annual salary of 50,000 dollar but includes an annual 5% performance bonus. The other job pays 55,000 dollor without any bonuses. Duc wants to know which job will provide him with a higher total income over 3 years if he performs well enough to receive the bonus.

Write a function that takes the salary, bonus rate, and number of years as inputs. Return the total income in a given number of year for Duc to decide






```python
def calculate_total_income(salary, bonus_rate, years):
    total_income_with_bonus = salary * (1 + bonus_rate / 100) * years
    return total_income_with_bonus

# Example usage
job1_income = calculate_total_income(50000, 5, 3)
job2_income = 55000 * 3

if job1_income > job2_income:
    print("Job 1 offers a higher total income.")
else:
    print("Job 2 offers a higher total income.")
```

**Question 4**


Duc is planning a trip and wants to decide on the mode of transportation based on the distance he needs to travel and the weather conditions. The rules for his decision are:

1.If the distance is less than 5 kilometers:  

  If it's not raining, Duc will walk.

  If it is raining, Duc will take a taxi.

2.If the distance is between 5 and 20 kilometers:

  If it's not raining, Duc will take a bicycle.

  If it is raining, Duc will take a bus.

3.If the distance is more than 20 kilometers:

  If it's not raining, Duc will drive a car.

  If it is raining, Duc will take a train.


Write a script using nested if statements to decide Duc's mode of transportation.


```python
def decide_transportation(distance, is_raining):
    if distance < 5:
        if not is_raining:
            return "walk"
        else:
            return "take a taxi"
    elif distance <= 20:
        if not is_raining:
            return "take a bicycle"
        else:
            return "take a bus"
    else:
        if not is_raining:
            return "drive a car"
        else:
            return "take a train"

# Example usage
print(decide_transportation(10, True))  # Output: take a bus
```
