1.1 Draw a box plot and give basic statistic information base on your observation

1.2 During the busiest time of the day at the Coffee Store, an average of 24 customers arrive per half-an-hour period.

Using the CDF(Cumulative distribution Function), answer these questions



b) What is the probability that the time between the arrivals of customers is 1 minute or more?

c) What is the probability that the next customer will arrive within 1 and 3 minutes?


```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = np.random.normal(0, 1, 100)  # Replace with your actual data

# Plotting the box plot
plt.figure(figsize=(10, 6))
plt.boxplot(data)
plt.title("Box Plot")
plt.ylabel("Values")
plt.show()

# Basic statistics
mean = np.mean(data)
median = np.median(data)
std_dev = np.std(data)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard Deviation: {std_dev}")
```

```python
import numpy as np

# Parameters
lambda_rate = 24 / 30  # Rate per minute

# Probability calculation
probability_1_minute_or_more = np.exp(-lambda_rate * 1)
print(f"Probability that time between arrivals is 1 minute or more: {probability_1_minute_or_more:.4f}")
```

```python
# Probability calculation
probability_within_1_and_3_minutes = (1 - np.exp(-lambda_rate * 3)) - (1 - np.exp(-lambda_rate * 1))
print(f"Probability that next customer arrives within 1 and 3 minutes: {probability_within_1_and_3_minutes:.4f}")
```

Run below code for the pie chart.
Please give 3 bullet points about the pitfalls of this chart from. What is missing?


And give plot a more suitable plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
categories = ['Category 3', 'Category 1', 'Category 2']
counts = np.array([450, 420, 200])

#pie chart

plt.figure(figsize=(8, 6))
plt.pie(counts, labels=categories, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart ")
plt.show()
```

Three pitfalls are:
1. Pie charts make it hard to compare the sizes of the slices accurately, especially when they are similar in size.
2. Categories are not ordered, which can make it challenging to quickly understand which category is larger or smaller.
3. While percentages are shown, it's difficult to grasp the exact numerical differences between categories without additional labeling.

```python
#your new chart goes here
import numpy as np
import matplotlib.pyplot as plt

# Data
categories = ['Category 3', 'Category 1', 'Category 2']
counts = np.array([450, 420, 200])

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(categories, counts, color=['blue', 'green', 'red'])
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Counts")
plt.show()
```
