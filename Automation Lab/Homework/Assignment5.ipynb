# Question 1:

Given the object is a right triangle with height is 4, base is 3. Create a  Triangle class to calculate the area and perimeter.

**hint: import math library and use the square root calculation math.sprt() to calculate the the hypotenuse**

```python
#your code goes here
import math

class Triangle:
    def __init__(self, height, base):
        self.height = height
        self.base = base

    def area(self):
        # Calculate the area of the triangle
        return 0.5 * self.height * self.base

    def hypotenuse(self):
        # Calculate the hypotenuse using the Pythagorean theorem
        return math.sqrt(self.height**2 + self.base**2)

    def perimeter(self):
        # Calculate the perimeter of the triangle
        return self.height + self.base + self.hypotenuse()

# Create an instance of Triangle with base 3 and height 4
triangle = Triangle(3, 4)
area = triangle.area()
perimeter = triangle.perimeter()
print(f"Area: {area}")
print(f"Perimeter: {perimeter}")
```

# Question 2:
Create a class CheckGrade to check for Duc grade in letter grade.

(" Duc Thanh Nguyen", 90)

Please check the given information, your code should print as expected output below

"This is Duc Thanh Nguyen, he got a/an A in Automation Lab."

```python
#your code goes here
class CheckGrade:
    def __init__(self, name, score):
        self.name = name
        self.score = score

    def get_letter_grade(self):
        # Determine the letter grade based on the score
        if self.score >= 90:
            return 'A'
        elif self.score >= 80:
            return 'B'
        elif self.score >= 70:
            return 'C'
        elif self.score >= 60:
            return 'D'
        else:
            return 'F'

# Instantiate the CheckGrade class for Duc Thanh Nguyen with a score of 90
student = CheckGrade("Duc Thanh Nguyen", 90)

# Get the letter grade
letter_grade = student.get_letter_grade()

# Print the expected output
print(f"This is {student.name}, he got a/an {letter_grade} in Automation Lab.")
```
