## Question 2

Create a Python function that allows the user to add a new product and its price to an existing dictionary of products. The function should prompt the user to type their input for both the product name and its price.

Then using the function to add
1.   "grape" : 2.5
2.   "apple" : 1.8

**Expected output example**

Enter the name of the product: Apple

Enter the price for Apple: 1.8

Product 'Apple' has been added/updated with price 1.8.


Updated Products List:

{'apple': 1.5, 'banana': 0.8, 'orange': 1.2, 'Grape': 2.5, 'Apple': 1.8}


```python
def add_product(products):
    product_name = input("Enter the name of the product: ")
    product_price = float(input(f"Enter the price for {product_name}: "))
    products[product_name] = product_price
    print(f"Product '{product_name}' has been added/updated with price {product_price}.")
    print("Updated Products List:")
    print(products)

products = {
    "apple": 1.5,
    "banana": 0.8,
    "orange": 1.2
}

add_product(products)
add_product(products)

```

## Question 3

Weather Data Analysis

Create a Python function to analyze temperature data for a week. The program should check if the data is empty or not, then compute the mean, minimum, maximum, and range of daily temperatures recorded in 10 days.




```python
def analyze_temperatures(data):
    if not data:
        print("No temperature data available.")
        return

    mean_temp = sum(data) / len(data)
    min_temp = min(data)
    max_temp = max(data)
    temp_range = max_temp - min_temp

    print(f"Mean Temperature: {mean_temp}")
    print(f"Minimum Temperature: {min_temp}")
    print(f"Maximum Temperature: {max_temp}")
    print(f"Temperature Range: {temp_range}")

# Example temperature data for a week
temperature_data = [12, 15, 12, 15, 19, 24, 26, 18, 21, 16]

# Analyze the temperature data
analyze_temperatures(temperature_data)

# #hint :these are som functions in python that you might use



```
