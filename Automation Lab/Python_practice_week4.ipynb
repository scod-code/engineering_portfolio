**Question 3**

The Exponential Distribution â€” A coffee store waiting time

During the busiest time of the day at the Coffee Store, an average of 24 customers arrive per half-an-hour period.

a) What is the probability that a customer will arrive within the next 2 minutes?

b) What is the probability that the time between the arrivals of customers is 1 minute or more?

c) What is the probability that the next customer will arrive within 1 and 3 minutes?

**3.1**

Generate 1,000 random samples representing the time between customer arrivals using the exponential distribution and plot the distribution

```python
customers_per_half_hour = 24  # customers per 30 minutes
time_period_minutes = 30  # time period in minutes

lambda_rate = customers_per_half_hour / time_period_minutes

# Generate random samples
samples = np.random.exponential(scale=1/lambda_rate, size=1000)

# Plot histogram of the samples
plt.hist(samples, bins=30, density=True, alpha=0.5, color='g')
plt.title('Histogram of Time Between Customer Arrivals')
plt.xlabel('Time (minutes)')
plt.ylabel('Density')
plt.show()
```

**3.2**

Calculate the PDF (Probability Density Function) from the generated data above and plot the function



```python
#pdf
x_values = np.linspace(0, max(samples), 1000)
pdf_values = lambda_rate * np.exp(-lambda_rate * x_values)

# Plot PDF
plt.figure()
plt.plot(x_values, pdf_values, 'r-', lw=2, label='PDF')
plt.hist(samples, bins=30, density=True, alpha=0.5, color='g', label='Histogram')
plt.title('PDF of Time Between Customer Arrivals')
plt.xlabel('Time (minutes)')
plt.ylabel('Density')
plt.legend()
plt.show()
```

**3.3**

Calculate the CDF(Cumulative distribution Function) from the generated data, plot the function and answer these question

a) What is the probability that a customer will arrive within the next 2 minutes?

b) What is the probability that the time between the arrivals of customers is 1 minute or more?

c) What is the probability that the next customer will arrive within 1 and 3 minutes?

```python
#cdf
sorted_samples = np.sort(samples)

cdf = 1- np.exp(-lambda_rate * sorted_samples)
# Plot CDF
plt.figure()
plt.plot(sorted_samples, cdf, 'b-', lw=2, label='CDF')
plt.title('CDF of Time Between Customer Arrivals')
plt.xlabel('Time (minutes)')
plt.ylabel('Cumulative Probability')
plt.legend()
plt.show()
```
