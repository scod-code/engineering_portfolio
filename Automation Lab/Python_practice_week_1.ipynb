#1. Slicing Lists

**list[start:stop:step]**

list[:: 1]



*   start: Not specified, so it defaults to the beginning of the list.

*   stop: Not specified, so it defaults to the end of the list.
*   step: =1 for 1 "step forward", = -1 for 1 "step backward"









```python

elements = ["orange", 25, 7.89, False, "mango", 200, True, "pineapple", 12.34, "watermelon"]
#use len() to check before slicing
print(len(elements))
# 1. Forward Slicing: Extract a subset from the beginning to the end of the list
print("Forward Slice_1:", elements[0:10])
print("Forward Slice_2:", elements[0:])
print("Forward Slice_3:", elements[::])
print("Forward Slice_4:", elements[::-1])


# 2. Backward Slicing: Extract a subset moving backward by using step -1
print("Backward Slice:", elements[-1:0:-1])

# 3. Slicing with Step -1: Reverse the entire list
print("Reversed List:", elements[::-1])

# 4. Slicing with Step: Extract every third element
print("Every Third Element:", elements[0::3])

# 5. Slicing with Colons for Subset: Extract a specific subset
print("Subset :", elements[2:8])
print("Subset with step 2:", elements[2:8:2])
```
