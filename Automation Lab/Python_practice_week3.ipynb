**Question1: Guessing Game**

Create a program that allows the user to guess a secret number between 1 and 100. The program should keep prompting the user until they guess the correct number.


```python
secret_number = 22
while True:
  guess = int(input("Guess a number between 1 and 100: "))
  if guess == secret_number:
    print("Congratulations! You guessed the correct number.")
    break
  else:
    print("Incorrect guess. Try again.")


# use while true so loop can run indefinitely until it is explicitly stopped

```

**Question 3**

Using zip to work with multiple iterables

Note: Stops when the shortest list ends



```python
#example
#As we can see, name is a shorter list, for loop will stop after 3 interations
names = ['Alice', 'Bob', 'Charlie']
ages = [25, 30, 35,33]

for name, age in zip(names, ages):
    print(f"{name} is {age} years old.")
```

Lists:

prices = [12, 18, 5, 7,8,2] (prices of the products)

quantities_sold = [10, 6, 15, 8,9,13] (quantities sold for each product)

Task:

Use zip to iterate over both lists.

For each product, calculate the revenue and print out revenue in this format

Product 1 Revenue: $120

```python
prices = [12, 18, 5, 7, 8, 2]
quantities_sold = [10, 6, 15, 8, 9, 13]

for i, (price, quantity) in enumerate(zip(prices, quantities_sold), start=1):
    revenue = price * quantity
    print(f"Product {i} Revenue: ${revenue}")
```

**Question 4**

Write a Python function and using for loop with condition statements to print the even numbers from a given list


given_list = [10, 21, 32, 43, 54, 65, 76, 87, 98]

```python
given_list = [10, 21, 32, 43, 54, 65, 76, 87, 98]

def print_even_numbers(numbers):
    for number in numbers:
        if number % 2 == 0:
            print(number)

print_even_numbers(given_list)
```

**Question 5**

You have a list of daily sales figures for a store over a week. Your task is to


1.   Create a for loop to calculate the total sales for the week.
2.   Create a second for loop to calculate the daily sales percentage contribution to the total sales


daily_sales = [150, 200, 175, 225, 190, 205, 210] (sales for 7 days)




```python
daily_sales = [150, 200, 175, 225, 190, 205, 210]

# Calculate total sales
total_sales = sum(daily_sales)
print(f"Total sales for the week: ${total_sales}")

# Calculate daily sales percentage contribution
for day, sales in enumerate(daily_sales, start=1):
    percentage = (sales / total_sales) * 100
    print(f"Day {day} sales contribution: {percentage:.2f}%")
```

**Question 6**

You are provided with a dictionary of student names and their grades. Write a program that:

Iterates through the dictionary using a for loop.

Assigns a letter grade based on the numerical grade using the following criteria:

A: 90 and above

B: 80-89

C: 70-79

D: 60-69

F: Below 60

Prints each student's name, their grade, and their letter grade.

```python
grades = {
    "Alice": 85,
    "Bob": 92,
    "Charlie": 58,
    "Diana": 60,
    "Eve": 72
}
```

```python
grades = {
    "Alice": 85,
    "Bob":92,
    "Charlie": 58,
    "Diana": 60,
    "Eve": 72
}

def assign_letter_grades(grades_dict):
    for student, grade in grades_dict.items():
        if grade >= 90:
            letter_grade = 'A'
        elif grade >= 80:
            letter_grade = 'B'
        elif grade >= 70:
            letter_grade = 'C'
        elif grade >= 60:
            letter_grade = 'D'
        else:
            letter_grade = 'F'

        print(f"{student}: Grade {grade}, Letter Grade {letter_grade}")

assign_letter_grades(grades)
```
