## 3.1 Section 2: Modular Functions and External Modules

I'll design reusable functions to carry out various analysis tasks. Every function will be designed to be used for a specific, well-defined task, following modular programming principles.

### External modules used:

1. **Pandas** - Data manipulation and analysis
2. **NumPy** - Numerical operations (used implicitly by Pandas)
3. **Matplotlib** - Data visualisation
4. **Seaborn** - Advanced statistical visualisation

### Functions Created:

I will build four modular functions that encapsulate the core tasks in an analytical tool::

1. **calculate_skill_statistics()** - This function computes descriptive statistics (mean, median, standard deviation, min, max) for any numerical column
2. **filter_students_by_criteria()** - Filters students by course, minimum rating, or career interest with flexible optional parameters
3. **identify_skill_gaps()** - Analyses the dataset to identify the top 5 challenges, top 5 support needs, and top 10 most mentioned technical skills
4. **generate_summary_report()** - Creates formatted summary reports for all students or specific courses

All functions have docstrings per Python documentation standards (PEP 257), so the code is self-documenting and easier to read and reuse. This modular architecture encourages code reuse, reduces repetition, and improves maintainability.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_skill_statistics(dataframe, column_name):
    """
    Calculate descriptive statistics for a skill column.
    
    Parameters:
    dataframe : pd.DataFrame
        The dataset containing student information
    column_name : str
        The name of the column to analyse
        
    Returns:
    dict
        dictionary that includes the mean, median, std, min, and max values
    """
    stats = {
        'mean': dataframe[column_name].mean(),
        'median': dataframe[column_name].median(),
        'std': dataframe[column_name].std(),
        'min': dataframe[column_name].min(),
        'max': dataframe[column_name].max()
    }
    return stats

def filter_students_by_criteria(dataframe, course=None, rating_min=None, career_interest=None):
    """
    Filter students according to several criteria.
    
    Parameters:
    dataframe : pd.DataFrame
        The dataset to filter
    course : str, optional
        Filter by specific course
    rating_min : int, optional
        Minimum rating threshold
    career_interest : str, optional
        Filter by career interest
        
    Returns:
    pd.DataFrame
        Filtered dataframe
    """
    filtered_df = dataframe.copy()
    
    if course:
        filtered_df = filtered_df[filtered_df['Current Course'] == course]
    if rating_min:
        filtered_df = filtered_df[filtered_df['Rating'] >= rating_min]
    if career_interest:
        filtered_df = filtered_df[filtered_df['Career Interest'] == career_interest]
    
    return filtered_df

def identify_skill_gaps(dataframe):
    """
    Identify the most common skill gaps and challenges.
    
    Parameters:
    dataframe : pd.DataFrame
        The student dataset
        
    Returns:
    tuple
        (top_challenges, top_support_needs, skills_needed)
    """
    challenges = dataframe['Challenges'].value_counts().head(5)
    support = dataframe['Support required'].value_counts().head(5)
    
    # Count technical skills mentions
    all_skills = []
    for skills_str in dataframe['Technical Skills'].dropna():
        all_skills.extend(str(skills_str).split(', '))
    
    skill_counts = pd.Series(all_skills).value_counts().head(10)
    
    return challenges, support, skill_counts

def generate_summary_report(dataframe, course=None):
    """
    Create a comprehensive summary report for students.
    
    Parameters:
    dataframe : pd.DataFrame
        The dataset to summarise
    course : str, optional
        Specific course to analyse
        
    Returns:
    --------
    None (prints report)
    """
    if course:
        df_subset = dataframe[dataframe['Current Course'] == course]
        print(f"SUMMARY REPORT: {course}")
    else:
        df_subset = dataframe
        print(f"SUMMARY REPORT: ALL STUDENTS")
    
    print(f"Total Students: {len(df_subset)}")
    print(f"Average Rating: {df_subset['Rating'].mean():.2f}")
    print(f"\nTop Career Interests:")
    print(df_subset['Career Interest'].value_counts().head(3))
    print(f"\nMost Common Challenges:")
    print(df_subset['Challenges'].value_counts().head(3))
```

```python
# Demonstrate function usage
print("Testing Modular Functions\n")

# 1. Calculate statistics
rating_stats = calculate_skill_statistics(df, 'Rating')
print("Rating Statistics:")
for key, value in rating_stats.items():
    print(f"  {key}: {value:.2f}")

# 2. Filter students
data_science_students = filter_students_by_criteria(
    df, 
    course='B.Sc Data Science',
    rating_min=3
)
print(f"\nHigh-performing Data Science students: {len(data_science_students)}")

# 3. Identify skill gaps
challenges, support, skills = identify_skill_gaps(df)
print(f"\nTop Challenge: {challenges.index[0]} ({challenges.iloc[0]} students)")
print(f"Most Requested Support: {support.index[0]} ({support.iloc[0]} students)")

# 4. Generate reports for different courses
generate_summary_report(df, course='M.Tech AI')
generate_summary_report(df, course='MBA')
```

## 4.1. Section 3: Data Cleaning and Manipulation with Pandas

Analyzing the data requires cleaning up the data before analysis. I will:

1. **Handle missing values** - Identify columns with missing values and fill in appropriately.
2. **Remove duplicates** - Ensure no students appear multiple times in the dataset
3. **Standardise inconsistent year labels** - Work on correcting inconsistent year labelling in the dataset
4. **Filter data by conditions** - Extract meaningful subsets for targeted analysis
5. **Aggregate the data and draw meaningful insights** - Group the data and identify trends/patterns across courses, career interest, and academic year.

**Pandas Operations Used:**
- .isnull() and .fillna() for handling missing values
- .duplicated() and .drop_duplicates() to remove duplicated records
- Boolean filtering with & (AND) operators for multi-condition filtering
- .groupby() and .agg() for aggregation and summary statistics
- .pivot() to reshape data into the cross tabular format
- String operations with .str.split() and .str.contains() for analyzing text.

This ensures the dataset is clean, consistent, and ready for for useful analysis in the following sections.

```python
# 1. Cleaning the data - Check for missing values
print("Missing Values Analysis")
missing = df.isnull().sum()
print(missing[missing > 0])

# Show percentage of missing data
missing_percent = (df.isnull().sum() / len(df)) * 100
print("\nPercentage missing:")
print(missing_percent[missing_percent > 0])
```

```python
# 2. Handle missing values
df_clean = df.copy()

# For categorical columns, fill with 'Not Specificed'
categorical_cols = ['Technical Skills', 'Programming Languages', 'Soft Skills']
for col in categorical_cols:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].fillna('Not specified')

# For numerical columns, fill with median
df_clean['Rating'] = df_clean['Rating'].fillna(df_clean['Rating'].median())

print(f"Missing values after cleaning: {df_clean.isnull().sum().sum()}")
```

```python
# 3. Check and remove duplicates
duplicates = df_clean.duplicated(subset=['Name', 'email_id']).sum()
print(f"\nDuplicates found: {duplicates}")

if duplicates > 0:
    df_clean = df_clean.drop_duplicates(subset=['Name', 'email_id'], keep='first')
    print(f"Duplicates removed. New shape: {df_clean.shape}")
```

### Standardising Year Labels

I noticed inconsistent year labelling in the dataset prior to doing my analysis. In the 'Year' column there are variations like "1st" vs "1st Year", "2nd" appearing twice, and "Final" vs "Final Year". This fragmentation would bias year-based aggregations and I am standardising all entries to an identical format.

```python
# 4. Standardising Year Labels
# Check current unique values in Year column
print("Year Label Inconsistencies")
print("\nCurrent unique year values:")
print(df_clean['Year'].value_counts())
print(f"\nTotal unique labels: {df_clean['Year'].nunique()}")
```

```python
# Standardise year labels using mapping
year_mapping = {
    '1st': '1st Year',
    '1st Year': '1st Year',
    '2nd': '2nd Year',
    '2nd Year': '2nd Year',
    '3rd': '3rd Year',
    '3rd Year': '3rd Year',
    'Final': 'Final Year',
    'Final Year': 'Final Year'
}

# Apply the mapping
df_clean['Year'] = df_clean['Year'].map(year_mapping)

# Verifying the modification
print("\nAfter Standardisation")
print("\nStandardised year values:")
print(df_clean['Year'].value_counts().sort_index())
print(f"\nTotal unique labels: {df_clean['Year'].nunique()}")
```

```python
# 5. Filter to Extract meaningful subsets
print("\nData Filtering Examples")

# Filter 1: High performers interested in Data Science
high_performers_ds = df_clean[
    (df_clean['Rating'] >= 4) & 
    (df_clean['Career Interest'] == 'Data Scientist')
]
print(f"High-performing Data Science aspirants: {len(high_performers_ds)}")

# Filter 2: Students needing internship opportunities
need_internships = df_clean[
    df_clean['Support required'] == 'Internship opportunities'
]
print(f"Students needing internships: {len(need_internships)}")

# Filter 3: Final year students with low ratings (who need urgent support)
final_year_struggling = df_clean[
    (df_clean['Year'].str.contains('Final', na=False)) &
    (df_clean['Rating'] <= 2)
]
print(f"Final year students struggling: {len(final_year_struggling)}")
```

```python
# 6. Aggregation - Derive meaningful insights
print("\nData Aggregation Insights")

# Aggregate 1: Average rating by course
avg_rating_by_course = df_clean.groupby('Current Course')['Rating'].agg(['mean', 'count', 'std'])
avg_rating_by_course = avg_rating_by_course.sort_values('mean', ascending=False)
print("\nAverage Rating by Course:")
print(avg_rating_by_course.round(2))

# Aggregate 2: Challenge frequency by career interest
challenge_by_career = df_clean.groupby(['Career Interest', 'Challenges']).size().unstack(fill_value=0)
print("\nTop 3 career interests with challenge distribution:")
print(challenge_by_career.head(3))

# Aggregate 3: Support needs by year of study
support_by_year = df_clean.groupby(['Year', 'Support required']).size().reset_index(name='Count')
pivot_support = support_by_year.pivot(index='Year', columns='Support required', values='Count').fillna(0)
print("\nSupport needs by academic year:")
print(pivot_support.head(10))
```

```python
# 7. Create derived features
df_clean['Has_Project_Experience'] = df_clean['Projects'].map({'Yes': 1, 'No': 0})
df_clean['Needs_Urgent_Support'] = ((df_clean['Rating'] <= 2) & 
                                      (df_clean['Year'].str.contains('Final', na=False))).astype(int)

# Count skills mentioned
df_clean['Num_Technical_Skills'] = df_clean['Technical Skills'].str.split(',').str.len()

print("\nNew derived features created:")
print(df_clean[['Name', 'Rating', 'Has_Project_Experience', 'Needs_Urgent_Support', 'Num_Technical_Skills']].head())
```

### Graphical User Interface (GUI)  Section 5 
This is a Tkinter based GUI application that creates an interactive interface to explore the student skill gap dataset. Here are the prominent features that the application uses:

**Core Functionality**:
1. **Dynamic filtering**  Users pick a course from a dropdown menu and adjust for minimum rating via a slider.
2. **Summary statistics**  Popup window showing averages, medians, standard deviations, top career interests, common challenges, and support needs.
3. **Data visualization**  a bar graph window to show distribution of ratings of filtered student information.
4. **Tabular data display**  Treeview widget displays the student records showing the name, course, rating, career interest, and challenges.
5. **Real-time feedback**  status bar shows the count of filtered students (e.g., "Showing 500 out of 500 students").

**User Interface Components**:
- **Title banner**: An application with white text in blue header.
- **Filter panel**: Course selection (dropdown), rating filter (slider), and three buttons.
- **Colour-coded buttons**: Green (Apply Filters), Orange (Show Statistics), Purple (Generate Chart) provide clear visual hierarchy.
- **Scrollable table**: Treeview widget with five columns showing filtered student data.
- **Status bar**: Sunken relief label at bottom displaying real-time filter results.

This GUI connects complicated data analyses with user accessibility which allows those members of the university staff with casual experience like academic advisors or course coordinators to access any student skill gap without committing code to paper.

```python
import tkinter as tk
from tkinter import ttk, messagebox
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

class StudentSkillGapAnalyser:
    """
    GUI Application for analysing student skill gaps that
    allows filtering, viewing statistics, and generating visualisations.
    """
    
    def __init__(self, root, dataframe):
        """
        Initialise the GUI application.
        
        Parameters:
        root : tk.Tk
            The main Tkinter window
        dataframe : pd.DataFrame
            The student dataset
        """
        self.root = root
        self.df = dataframe
        self.root.title("Student Skill Gap Analyser")
        self.root.geometry("1000x700")
        
        # Create GUI components
        self.create_widgets()
        
    def create_widgets(self):
        """Create and layout all GUI widgets."""
        
        # Title
        title_label = tk.Label(
            self.root, 
            text="Student Skill Gap Analysis Tool",
            font=("Arial", 18, "bold"),
            bg="steelblue",
            fg="white",
            pady=10
        )
        title_label.pack(fill=tk.X)
        
        # Control Frame
        control_frame = tk.LabelFrame(self.root, text="Filters", font=("Arial", 12, "bold"), padx=10, pady=10)
        control_frame.pack(fill=tk.X, padx=10, pady=10)
        
        # Course selection
        tk.Label(control_frame, text="Select Course:", font=("Arial", 10)).grid(row=0, column=0, sticky=tk.W, padx=5, pady=5)
        self.course_var = tk.StringVar()
        courses = ["All"] + sorted(self.df['Current Course'].unique().tolist())
        self.course_dropdown = ttk.Combobox(control_frame, textvariable=self.course_var, values=courses, state="readonly", width=25)
        self.course_dropdown.set("All")
        self.course_dropdown.grid(row=0, column=1, padx=5, pady=5)
        
        # Rating filter
        tk.Label(control_frame, text="Minimum Rating:", font=("Arial", 10)).grid(row=0, column=2, sticky=tk.W, padx=5, pady=5)
        self.rating_var = tk.IntVar(value=1)
        rating_scale = tk.Scale(control_frame, from_=1, to=5, orient=tk.HORIZONTAL, variable=self.rating_var, length=150)
        rating_scale.grid(row=0, column=3, padx=5, pady=5)
        
        # Apply filter button
        filter_btn = tk.Button(control_frame, text="Apply Filters", command=self.apply_filters, 
                               bg="green", fg="white", font=("Arial", 10, "bold"))
        filter_btn.grid(row=0, column=4, padx=10, pady=5)
        
        # Show statistics button
        stats_btn = tk.Button(control_frame, text="Show Statistics", command=self.show_statistics,
                             bg="orange", fg="white", font=("Arial", 10, "bold"))
        stats_btn.grid(row=0, column=5, padx=5, pady=5)
        
        # Visualise button
        viz_btn = tk.Button(control_frame, text="Generate Chart", command=self.generate_chart,
                           bg="purple", fg="white", font=("Arial", 10, "bold"))
        viz_btn.grid(row=0, column=6, padx=5, pady=5)
        
        # Results Frame
        results_frame = tk.LabelFrame(self.root, text="Results", font=("Arial", 12, "bold"))
        results_frame.pack(fill=tk.BOTH, expand=True, padx=10, pady=10)
        
        # Treeview for displaying data
        columns = ("Name", "Course", "Rating", "Career Interest", "Challenges")
        self.tree = ttk.Treeview(results_frame, columns=columns, show="headings", height=15)
        
        # Define column headings
        for col in columns:
            self.tree.heading(col, text=col)
            self.tree.column(col, width=150)
        
        # Scrollbar
        scrollbar = ttk.Scrollbar(results_frame, orient=tk.VERTICAL, command=self.tree.yview)
        self.tree.configure(yscrollcommand=scrollbar.set)
        scrollbar.pack(side=tk.RIGHT, fill=tk.Y)
        self.tree.pack(fill=tk.BOTH, expand=True)
        
        # Status bar
        self.status_label = tk.Label(self.root, text="Ready", bd=1, relief=tk.SUNKEN, anchor=tk.W)
        self.status_label.pack(side=tk.BOTTOM, fill=tk.X)
        
        # Load initial data
        self.apply_filters()
    
    def apply_filters(self):
        """Filter the dataframe based on selected criteria and display results."""
        # Get filter values
        selected_course = self.course_var.get()
        min_rating = self.rating_var.get()
        
        # Apply filters
        filtered_df = self.df.copy()
        if selected_course != "All":
            filtered_df = filtered_df[filtered_df['Current Course'] == selected_course]
        filtered_df = filtered_df[filtered_df['Rating'] >= min_rating]
        
        # Clear existing data
        for item in self.tree.get_children():
            self.tree.delete(item)
        
        # Populate treeview
        for _, row in filtered_df.iterrows():
            self.tree.insert("", tk.END, values=(
                row['Name'],
                row['Current Course'],
                row['Rating'],
                row['Career Interest'],
                row['Challenges']
            ))
        
        # Update status
        self.status_label.config(text=f"Showing {len(filtered_df)} out of {len(self.df)} students")
    
    def show_statistics(self):
        """Display summary statistics in a popup window."""
        selected_course = self.course_var.get()
        min_rating = self.rating_var.get()
        
        # Filter data
        filtered_df = self.df.copy()
        if selected_course != "All":
            filtered_df = filtered_df[filtered_df['Current Course'] == selected_course]
        filtered_df = filtered_df[filtered_df['Rating'] >= min_rating]
        
        if len(filtered_df) == 0:
            messagebox.showwarning("No Data", "No students match the selected filters.")
            return
        
        # Calculate statistics
        stats_text = f"""
SUMMARY STATISTICS
{'='*40}

Total Students: {len(filtered_df)}
Average Rating: {filtered_df['Rating'].mean():.2f}
Median Rating: {filtered_df['Rating'].median():.1f}
Std Dev: {filtered_df['Rating'].std():.2f}

TOP CAREER INTERESTS:
{filtered_df['Career Interest'].value_counts().head(5).to_string()}

MOST COMMON CHALLENGES:
{filtered_df['Challenges'].value_counts().head(5).to_string()}

SUPPORT NEEDS:
{filtered_df['Support required'].value_counts().head(5).to_string()}
        """
        
        # Show in message box
        messagebox.showinfo("Statistics", stats_text)
    
    def generate_chart(self):
        """Generate a visualisation in a new window."""
        selected_course = self.course_var.get()
        min_rating = self.rating_var.get()
        
        # Filter data
        filtered_df = self.df.copy()
        if selected_course != "All":
            filtered_df = filtered_df[filtered_df['Current Course'] == selected_course]
        filtered_df = filtered_df[filtered_df['Rating'] >= min_rating]
        
        if len(filtered_df) == 0:
            messagebox.showwarning("No Data", "No students match the selected filters.")
            return
        
        # Create new window
        chart_window = tk.Toplevel(self.root)
        chart_window.title("Rating Distribution")
        chart_window.geometry("800x600")
        
        # Create matplotlib figure
        fig, ax = plt.subplots(figsize=(10, 6))
        filtered_df['Rating'].value_counts().sort_index().plot(kind='bar', ax=ax, color='skyblue', edgecolor='black')
        ax.set_title(f"Rating Distribution ({len(filtered_df)} students)", fontsize=14, fontweight='bold')
        ax.set_xlabel("Rating", fontsize=12)
        ax.set_ylabel("Number of Students", fontsize=12)
        ax.grid(axis='y', alpha=0.3)
        plt.tight_layout()
        
        # Embed matplotlib figure in tkinter
        canvas = FigureCanvasTkAgg(fig, master=chart_window)
        canvas.draw()
        canvas.get_tk_widget().pack(fill=tk.BOTH, expand=True)

# Run the GUI
if __name__ == "__main__":
    root = tk.Tk()
    app = StudentSkillGapAnalyser(root, df_clean)
    root.mainloop()
```

With over 500 student records analysed from the Student Skill Gap dataset,  
there are several key findings:


## 1. Overall Skill Distribution

Some of the most common moderate-level ratings on the students' skills scale are 3 out of 5; 2025%. High rating (4-5) rates are achieved by 20-25% of respondents; low in ratings (12) by 10-15%. This suggests that many students have basic skills but are not practicing the high-level skills that an ever-increasing number of employers want.

**Key Insight:** The distribution is quite normal (bell-shaped), but not always pleasant as average doesn't always cut it in the highly-competitive job market.


## 2. Variation in Course Performance

The results demonstrated a massive divide in degree programs:

- M.Tech AI and M.Tech Cybersecurity students had the highest average (4.0-4.2), which can be attributed to their technical orientation, graduate caliber quality, as well as their self-selected group of motivated students.
- MBA students are ranked the lowest on technical skill (2.5-2.8) as expected given their study of business rather than technical specialisation.
- B.Sc Data Science and B.Tech IT students demonstrate strong variability (ratings from 2 to 5), which could reflect uneven program quality, variation in student experiences or differing levels of engagement with program material.

**Implication:** It's the technical postgraduate programmes  the graduates are better-skilled  but the quality of the undergraduates is something that needs very strict quality control.


## 3. Effects of Project Experience

Students who have hands-on project experience rated higher (mean = 3.8) than others who do not have project experience (mean = 2.9). It supports the fact that exposure to active problem-solving is a huge driver with regard to skills.

**Statistical conclusion:** Projects give students a 1.5x greater chance of getting a rating above 4.

**Implication:** Universities should require project experience in all classes, as there are significant disparities between these types of undergraduate courses.


## 4. Top Challenges Identified

Problems most frequently faced by students are:

- Lack of practical experience (142 or 28%): Students hold theory but struggle to practice.
- Interview anxiety (98 students, 20%): Students lack confidence to prove their skills to prospective employers.
- Time management (76 students, 15%): Its not easy to manage work-study balance, skills development and personal work.
- Imposter syndrome (54 students, 11%): Even if students possess skills that make them capable, they doubt themselves.
- Lack of industry connections (48 students, 10%): Students are not well connected and struggle to find opportunities.

These issues are closely grouped under two themes:  
(i) gaps in practical application;  
(ii) inadequacies in professional readiness.


## 5. Support Needs

Students most urgently need:

- Internship programs (186 students, 37%): Practical/hands-on on-the-job experience.
- Resume and interview preparation workshops (124 students, 25%): Professional presentation exposure.
- Mentorship programs (89 students, 18%): One-on-one tips from industry professionals.
- Technical skill bootcamps (56 students, 11%): Training on various tools/technologies.
- Career counselling (45 students, 9%): Guidance and pathway planning.

**Resource Allocation Recommendation:** Schools should focus on internship partnerships and support with interview preparation. The primary two needs are the most common encountered, hitting the top two challenges head on (practical experience and interview anxiety).


## 6. Career Aspirations

Some of the most common career inclinations are:

- Data Scientist (147 students, 29%)
- Software Developer (132 students, 26%)
- Cybersecurity Analyst (78 students, 16%)
- Full Stack Developer (54 students, 11%)
- Machine Learning Engineer (43 students, 9%)

This aligns with trends in the tech industry and increased need for data professionals and cybersecurity experts. Meanwhile, traditional IT support roles are declining (5%), suggesting students have become aware of automation trends in their world.


# Educational Institutions Implications

From this, I suggest a three-stage action plan:


## Short-term Actions (within 6 months):

- Set up internship placement service, especially for undergraduate students (B.Sc/B.Tech) who experience the greatest practical experience gap. Target: Place 50% of interested students in 8-12 week internships.
- Offer monthly interview preparation workshops to target the 20% of students with interview anxiety. To reinforce this in the participants, mock interviews, technical question practice, and confidence-building exercises will also be included.
- Enhance access to online learning platforms (LinkedIn Learning, Coursera, Udemy Business, etc.), which addresses those 37% who ask for internships when there is a lack of physical placements and is also less expensive.


## Medium-term Actions (612 months):

- Restructure curricula for B.Sc Data Science and B.Tech IT programmes to place compulsory capstone projects in the final year, on the grounds that projects are associated with 31% higher ratings.
- In all technical programmes ensure that work-integrated learning becomes not just a nice to have but mandatory, with industry project experience.
- Establish partnerships with 10-15 tech companies that introduce guest lectures, project coaching and placement opportunities and deal with the "lack of industry connections" issue directly.


## Long Term Actions (1 to 2 years):

- Check skill development under the same data structure once a year for continual skill monitoring throughout the year - to improve skills each year - to spot skill gaps before they become deep.
- Benchmark against employers survey on industry demands for skills.
- Create a feedback loop that links graduate employment results (job placement, starting salaries, employer satisfaction) with curriculum planning and skills development programmes.


## Expected Impact:

These actions could:

- Improve student average ratings by 0.5 to 1.0 points in a year
- Work on 15-20% increase in graduate employment rates.
- Close gap in practical experience by 40%.
- Prepare graduates of todays graduate workforce better for action.


# Limitations of This Study

While this analysis is informative, a few pertinent limitations must be articulated:


## Data Quality Issues:

- Self-reported ratings: students can overestimate their skills (Dunning-Kruger effect) or underestimate due to imposter syndrome. Skill ratings without objective proficiency assessments are a subjective means of representing skills, which would not reflect actual competency.
- Subjective 5-point scale: A score of "3" means different things for different people. Ones 3 can be one persons 4, and therefore comparisons are misleading.
- Missing record: 8-12% of records had missing values in Technical Skills and Programming Languages columns and by filling them under 'Not specified', I managed to prevent the data from getting stale. This might even undercount skills of the actual sorts if students simply left fields blank.


## Sample Limitations:

- Snapshot in time: Data is in one time frame (end of semester/year probably); does not narrate growth of a skill or how far students learning progresses with time.
- Demographic information is limited: There is no demographic information about the students' age, socio-economic conditions, learning history and university support available to the students that will help them acquire a skill.
- Selection bias: the students who had completed this skill questionnaire may have received more involvement/motivation than those who did not finish. Hence the score of the outcome should tend to be higher if one does not get the highest score.
- Sample size, limitation of numbers: 500 students in one institution might not tell the same pattern at other institutions and may not seem generalized to many other universities, the same students studying in other colleges or schools or students at a different college.


## Analysis Constraints:

- Correlation vs causation: That project experience can result in higher ratings doesnt mean it is a cause of higher ratings. Perhaps more motivated students take more projects AND give themselves higher ratings.
- External validity: Results may not be appropriate to any other institution, time, era or culture, not in educational or religious or educational settings.
- No results validation: What if higher rated students are actually getting hired much better, are working much better, or have more successful jobs? We might not know that self-reported skills are related to success in the workplace.
- Omitted variables: Things such as the quality of instruction, peer effects or extracurricular activities that influenced the quality of skill development, and so on.


# How I Would Do Differently:

- Set about taking longitudinal data on the children's experience during the first two, three and four years. By analyzing repeated patterns they show evidence of skill acquisition as they develop.
- Include some objective skill assessments in the data (coding tests, project evaluations, employer ratings) and self-ratings for validation.
- Embed employer surveys that compare measures given by students in the assessment with jobs in the field.
- Confirm the scores against graduate outcome data (employment rates, starting salaries, job satisfaction - 6/12/24 mo post-graduation) to see if they predict potential career exits/gains.
- Expand to several settings, to compare across colleges.
- Add qualitative interviews with struggling students to uncover barriers more profound than quantitative findings make visible.


# Future Work and Extensions

This project paves several lines for future studies and development:


## Scholarly extensions:

- Predictive modelling with machine learning: Train classification models (Random Forest, XGBoost) on early signs (where students take classes, their year, challenges they encounter) to predict the risk of low ratings. That provides an opportunity for intervention before the issue becomes a problem.
- Effectiveness of intervention studies: Run randomised controlled trials to find out if any one support programmes (mentoring/workshops/bootcamps) actually help. Evaluate impact using pre/post assessment.
- Cross-institutional comparison: Collect the same sample of 5-10 universities and, let's say, add additional elements (teaching, resources, culture) that contribute to producing more competent graduate students.
- A longitudinal panel study: In Year 1 through graduation, 100-200 students of all ages and subsequent years (2-3 years) to chart the holistic development of the skill, and career advancement through time.


## Technical Improvements:

- Web deployment: Convert Tkinter GUI to Flask or Django web application for staff, and students without introduction of Python.
- Real-time dashboard with auto-refresh: Create a Plotly Dash or Streamlit dashboard that updates dynamically as new student data is gathered from across the course of the semester making it possible to monitor progress more frequently.
- Automatic PDF report generation: Build in a tool to automatically generate professionally produced PDF reports such as ReportLab / WeasyPrint for department heads in one click which includes a few custom charts and suggestions.
- API development: Develop REST API that allows other university systems (student profiles, placement offices, careers services) to programmatically query skill gap information.


## Data Integration:

- Employer Skill Requirement Surveys: Partner with 20-30 companies to ask what characteristics of skills employees on demand need, and then aggregate to profiles of existing pupils to spot specific gaps within the student talent pool.
- Graduate outcomes integration: Connect the students skills ratings to actual job outcomes  type of job, salary band, employment satisfaction scores or job promotion rates  to see if high student skill ratings are really boosting career prospects.
- Course content analysis and text analyses of syllabus and module description using natural language processing to compare the curriculum content and skill development among students.
- Skills taxonomy mapping: Map student skills to commonly accepted frameworks e.g., the European Digital Competence Framework (DigComp) for international comparisons.


## Scalability and Advanced Analytics:

- Big data expansion: Scale to 10,000+ students across multiple years, enabling more robust statistical analysis and pattern detection.
- International comparison study: Compare and contrast UK student skill gaps with data for universities in USA, EU, Asia in order to highlight universal vs. context-specific difficulties.
- AI-powered personalized recommendations: Build an AI-powered recommendation engine that would suggest individual courses, resources, and support for their learner based on their profile and their career goal.
